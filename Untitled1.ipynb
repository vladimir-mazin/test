Both LiF and KBr have face-centered cubic crystal lattice with the lattice constant equal to 0.403 and 0.659 nm correspondingly. The extinction rules for this type of lattice are : h, k, l all odd or all even, thus the first observable diffraction peaks have (200) indces.

LiF crystal

coupling mode	1 :2

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

Load measurement data

```python
filename = 'lif.txt'
_, rate, _, _, _, theta = np.genfromtxt(filename, unpack=True, skip_header=3)
```

```python
plt.plot(theta, rate)
#plt.yscale('log')
plt.grid()
```

```python
peaks,_=find_peaks(rate, height=0)
plt.plot(peaks, rate[peaks],"x")
```
